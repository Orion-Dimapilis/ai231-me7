```python
import cv2
from transformers import AutoProcessor, BlipForConditionalGeneration
from PIL import Image
import torch

# Load processor and model for BLIP image captioning
processor = AutoProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")

# Check if GPU is available and move model to GPU if so
device = "cuda" if torch.cuda.is_available() else "cpu"
model = model.to(device)

# Open webcam (0 is default for the first webcam)
cap = cv2.VideoCapture(2)

# Define your question (you can change this to different prompts)
question = "What can you see?"

# Process webcam frames continuously
while True:
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame")
        break

    # Convert the frame to a PIL Image for the model
    image = Image.fromarray(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))

    # Preprocess the image and question
    inputs = processor(images=image, text=question, return_tensors="pt").to(device)

    # Generate the answer
    outputs = model.generate(**inputs)
    
    # Decode the answer
    answer = processor.decode(outputs[0], skip_special_tokens=True)

    # Debug: print the model's output before displaying it
    print(f"Model Output: {answer}")
    
    # Show the answer on the webcam feed
    cv2.putText(frame, f"Answer: {answer}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)
    
    # Display the frame with the generated answer
    cv2.imshow("Webcam", frame)

    # Exit the loop on pressing 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```
